```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_csv(r"C:\Users\45237\OneDrive - Danmarks Tekniske Universitet\Skrivebord\assignment\assignment\task_2\HR_data.csv")
df_round1 = df[df['Round'] == 'round_1'] # for checking rounds seperately...

# List of HR features and target variable
features = ['HR_Mean', 'HR_Median', 'HR_std', 'HR_Min', 'HR_Max', 'HR_AUC']
target = 'Frustrated'

# Set up a grid of subplots
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.flatten()

# Create scatter plots with linear regression lines
for i, feature in enumerate(features):
    sns.regplot(
        data=df,
        x=feature,
        y=target,
        ax=axes[i],
        scatter_kws={'s': 30},        # Size of points
        line_kws={'color': 'red'}     # Line color
    )
    axes[i].set_title(f'{feature} vs {target}', fontsize =17)
    axes[i].set_xlabel(feature, fontsize=15)
    axes[i].set_ylabel(target, fontsize=15)

plt.tight_layout()
plt.show()
```
